```python
"""
Docstring:
Introduction : It is a gameshow to test how good is your general knowledge?. This game consists of 5 stages.
Stage 1: Who is the fastest round?
Stage 2: Introduction to the actual game and option of using lifelines.
Stage 3: Two questions will be introduced which will have the fifth option of taking money home.
Stage 4: Last Question, this question will not have any lifelines and fifth option of taking money home.
Stage 5: Favour round.
To help participant, there are "lifelines" that will let them either eliminate two of the four possible answers or take a 
poll of the audience for what they think the right answer is.
Only one lifeline per question can be used (Maximum 2 lines). Moreover, for stage 3 if the participant uses lifeline he won't be allowed
to use the fifth option of taking money home.
Host for the game will be Computer.
"""

from sys import exit
import random
from matplotlib import pyplot as plt
%matplotlib inline
import time


prize1=500
prize2=1000
prize3=1500
prize4=2000

fif_50_var=1
AudienceP_var=1
count_lifeline=fif_50_var + AudienceP_var
participant_name ="" 

#This functions shows bar graph for Audience poll lifeline
def Audience(options):
    """This functions shows bar graph for Audience poll lifeline"""
    global count_lifeline
    global AudienceP_var
    count_lifeline=count_lifeline-1
    AudienceP_var=0
    x = options
    if len(options)>4:  #Checking if the options are greater than 4 if yes then remove last one
        options.pop(4)
        x=options
    num_of_votes = random.sample(range(1, 100), 4)
    #while loop so that (num_of_votes total=100) when we take randomly 4 numbers their sum should total to 100
    while sum(num_of_votes) != 100:
        num_of_votes = random.sample(range(1, 100), 4)
    max_vote= max(num_of_votes)
    bar1=plt.bar(x, num_of_votes, align='center', alpha=0.5) #plotting the graph   
    #for loop to find maximum number of votes for an option
    for j in range(len(x)):
        if num_of_votes[j]==max_vote:
            labell=x[j]
    plt.ylabel('Votes')
    plt.xlabel('Options')
    
    for i in bar1: #To show values on the top of bar graph
        height = i.get_height()
        plt.text(i.get_x()+ i.get_width()/2.0, height, s=int(height), horizontalalignment='center', verticalalignment='bottom')
    plt.tight_layout()
    plt.show()
    print("Maximum number of people voted for option: " + str(labell))

#This function eliminates two options from the list of four
def fifty50(option,answerr):
    """This function eliminates two options from the list of four"""
    global count_lifeline 
    global fif_50_var
    count_lifeline=count_lifeline-1
    fif_50_var=0
    answer_list=answerr   #Saving the right answer in variable
    if len(option)>4:
        option.pop(4)
    option.remove(answerr)    #Removing the right answer from original list
    r1=option.pop(random.randrange(len(option))) #to randomly pop out an answer
    r2=option.pop(random.randrange(len(option)))
    option.append(answer_list) #Appending the right answer to the list from where randomly answers were deleted
    option.sort() #Sorting the new list
    return option    

def fail():
    """This function is called when user loses the game"""
    print(f"""
    
            Because your answer is wrong... In the end it was a good game. Better luck next time {participant_name} 
            
        """)
    print(f"""
                                        Thanks for playing {participant_name}
        """)
def congrats():
    print("""                                 
                                    

_________                                     __        .__          __  .__                      
\_   ___ \  ____   ____    ________________ _/  |_ __ __|  | _____ _/  |_|__| ____   ____   ______
/    \  \/ /  _ \ /    \  / ___\_  __ \__  \\   __\  |  \  | \__  \\   __\  |/  _ \ /    \ /  ___/
\     \___(  <_> )   |  \/ /_/  >  | \// __ \|  | |  |  /  |__/ __ \|  | |  (  <_> )   |  \\___ \ 
 \______  /\____/|___|  /\___  /|__|  (____  /__| |____/|____(____  /__| |__|\____/|___|  /____  >
        \/            \//_____/            \/                     \/                    \/     \/ 

""")

#Function to check which lifeline is used
def rep(answer,life_l,not_blank,questn1,right_answers,questionnumber) :
    """Function to check which lifeline is used"""
    #if to check the count of lifeline
    if count_lifeline > 0 :
        #while loop will run till user inputs a valid input
        while life_l!="1" or life_l !="2" or life_l!="3" or life_l.lower()=="one" or life_l.lower()=="two" or life_l.lower()=="three":   
            if not_blank!="1":
                #to check if fifty/fifty lifeline is not available then print only one life line available Audience poll
                if fif_50_var == 0:
                    print('-' * 40)
                    life_l=input(f"\nDo you want to use lifeline for this question?\nYou have only one left \nType 2 for Audience Poll \n3 for No\n{participant_name} you want to use lifeline?")
                    print('-' * 40)
                #to check if Audience poll lifeline is not available then print only one life line available fifty/fifty
                if AudienceP_var==0:
                    life_l=input(f"\nDo you want to use lifeline for this question?\nYou have only one left\nType 1 for 50/50 \n 3 for No\n{participant_name} you want to use lifeline?") 
                #to check if both lifelines available then print there are two lifelines
                if fif_50_var >0 and AudienceP_var >0 :
                        print("-" * 40 +"\nDo you want to use lifeline for this question?\nTYPE 1 for 50/50 \n 2 for Audience Poll \n 3 for No")
                        print('-' * 40)
                        life_l=input(f"{participant_name} you want to use lifeline?")
            #To check which lifeline user selected
            if life_l == "1" or life_l =="2" or life_l.lower()=="one" or life_l.lower()=="two" or life_l =="50/50" or life_l =="50\50" or life_l.lower()=="audience poll"  :
                if (life_l=="1" or life_l.lower()=="one" or life_l =="50/50" or life_l =="50\50") and fif_50_var > 0:
                    new_question1_options=fifty50(questn1,right_answers)
                    print("Your new options are \n")
                    print(*new_question1_options[0:], sep = "\n")
                    answer =input("Press the number besides the option which you think is right\n")
                elif (life_l=="2" or life_l.lower()=="two" or life_l.lower()=="audience poll") and AudienceP_var >0:
                    Audience(questn1)
                    answer =input("Press the number besides the option which you think is right\n")
                else: #this else: if the user inputs 50/50 lifeline(option 1) even when its not available to him
                      #this else statemennt will again ask user to input the value
                    if fif_50_var == 0:
                        print('-' * 40)
                        print("You don't have that lifeline")
                        print('-' * 40)
                        while ((life_l=="1" or life_l.lower()=="one" or life_l =="50/50" or life_l =="50\50") or ((life_l!="3" and life_l.lower()!="three" and life_l.lower()!="no") and (life_l!="2" and life_l.lower()!="two" and life_l.lower() !="audience poll")))  :
                            print('-' * 40)
                            life_l=input(f"Do you want to use lifelines?\nYou have only one left \nType 2 for Audience Poll \n3 for No\n{participant_name} you want to use lifeline?")
                            print('-' * 40)
                        if (life_l=="2" or life_l.lower()=="two" or life_l.lower()=="audience poll"):
                            Audience(questn1)
                            answer =input("Press the number besides the option which you think is right\n")
                        elif life_l=="3" or life_l.lower()=="three" or life_l.lower()=="no":
                            answer =input("Press the number besides the option which you think is right\n")
                    elif AudienceP_var==0:
                        print('-' * 40)
                        print("You don't have that lifeline")
                        print('-' * 40)
                        while ((life_l=="2" or life_l.lower()=="two" or life_l.lower() =="audience poll") or ((life_l!="3" and life_l.lower()!="three" and life_l.lower()!="no") and (life_l!="1" and life_l.lower()!="one" and life_l !="50/50" and life_l !="50\50")))   :
                            print('-' * 40)
                            life_l=input(f"Do you want to use lifelines?\nYou have only one left \nType 1 for 50/50 \n3 for No\n{participant_name} you want to use lifeline?")
                            print('-' * 40)
                        if (life_l.lower()=="one" or life_l=="1" or life_l=="50/50" or life_l=="50\50"):
                            new_question1_options=fifty50(questn1,right_answers)
                            print("Your new options are \n")
                            print(*new_question1_options[0:], sep = "\n")
                            answer =input("Press the number besides the option which you think is right\n")
                        elif life_l=="3" or life_l.lower()=="three" or life_l.lower()=="no":
                            answer =input("Press the number besides the option which you think is right\n")
                        
            elif life_l=="3" or life_l.lower()=="three" or life_l.lower()=="no":
                answer =input("Press the number besides the option which you think is right\n")
            if answer!= None :
                answer= check_answers_inlist(questionnumber,answer)
                return(answer)            
            else: 
                print('-' * 40)
                life_l=input(f"\nPlease enter one of these \n Do you want to use lifelines? Press 1 for 50/50 \nPress 2 for Audience Poll \nPress 3 for No\n{participant_name} you want to use lifeline?")
                print('-' * 40)
                not_blank="1"                
    else:
        answer =input("Press the number besides the option which you think is right\n")
        if answer!= None :
            answer= check_answers_inlist(questionnumber,answer)
            return(answer) 
        

#Function used to handle invalid user input when he/she is answering the questions(Question1,Question2,Question3,Last Question)
def check_answers_inlist(questionnum,answervar): 
    """Function used to handle invalid user input when he/she is answering the questions(Question1,Question2,Question3,Last Question)"""
    count=0
    while answervar != "":
        if questionnum=="1" :
            options=["60", "30", "83.33", "0.012" ]
            numbers=["1","2","3","4"]
            options_in_words=["Sixty","Thirty","Eighty three","point zero one two"]
            num_word=['one','two','three','four']
            lower_case=list(map(lambda l:l.lower(),options_in_words))  
            for (a,b,c,d) in zip(options,num_word,lower_case,numbers) :                
                if answervar == a:
                    count=1
                    return(answervar)        
                if answervar.lower() == b:
                    count=1
                    return(answervar)
                if answervar.lower() == c:
                    count=1
                    return(answervar)
                if answervar ==d:
                    count=1
                    return(answervar)
        if questionnum=="2" :
            options=["1945", "1920", "1914", "1898", "Walk away with $100 (If you guess incorrectly, you walk away with nothing)"]
            numbers=["1","2","3","4","5"]
            num_word=['one','two','three','four','five'] 
            for (a,b,c) in zip(options,num_word,numbers) :
                count=0
                if answervar == a:
                    count=1
                    return(answervar)        
                if answervar.lower() == b:
                    count=1
                    return(answervar)
                if answervar== c:
                    count=1
                    return(answervar)
        if questionnum=="3" :
            options=["1", "2", "3", "4" ,"5" ]
            options_in_words=["Barber","Knight","Blacksmith","Jester","Walk away"]
            num_word=['one','two','three','four','five']
            lower_case=list(map(lambda l:l.lower(),options_in_words))  
            for (a,b,c) in zip(options,num_word,lower_case) :
                if answervar == a :
                    count=1
                    return(answervar)        
                if answervar.lower() == b:
                    count=1
                    return(answervar)
                if  answervar.lower() == c:
                    count=1
                    return(answervar)
        if questionnum=="4" :
            options=["1", "2", "3", "4" ]
            options_in_words=["Moth","Roach","Fly","Beetle"]
            num_word=['one','two','three','four']
            lower_case=list(map(lambda l:l.lower(),options_in_words))  
            for (a,b,c) in zip(options,num_word,lower_case) :
                if answervar == a :
                    count=1
                    return(answervar)        
                if answervar.lower() == b:
                    count=1
                    return(answervar)
                if  answervar.lower() == c:
                    count=1
                    return(answervar)
        if count==0:
             answervar=input("Invalid output.. Please answer again")

#setting common variable in multiple functions to a value                
def set_var():
    """setting common variable in multiple functions to a value """
    global answer
    global life_lines
    global not_blanks
    answer = None   
    life_lines=None
    not_blanks="0"
    
#Function to transfer call from one function to another
def to_continue_play(numques,continue_play):
    """Function to transfer call from one function to another"""
    
    while (continue_play !="1" and continue_play !="2") and (continue_play.lower()!="one" and continue_play.lower()!="two") and (continue_play.lower()!="yes" and continue_play.lower()!="no"):
        continue_play=input("Please enter one of the following: Do you want to continue playing?\nTYPE 1 for Yes \n2 for No ")
    if continue_play == "1" or continue_play.lower()=="one" or continue_play.lower()=="yes":
        if numques=="1":
                print("""
                                            Lets Proceed to the Second Question
                """)
                time.sleep(3)
                question2()  
        elif numques=="2":
                print("""
                                            Lets Proceed to the Third Question
                """)
                time.sleep(3)
                question3() 
        elif numques=="3":
                print("""
                                            Lets Proceed to the Last Question
                """)
                time.sleep(3)
                questionLast() 
        elif numques=="4":
                time.sleep(3)
        else:
                print("Invalid")
    else:
        if numques=="1":
                print("""
                                        Bye!! You go home with $ """ + str(prize1))  
        elif numques=="2":
                print(f"""
                                Bye!! You go home with $ {prize1}. Don't, uh, spend it all at once...""")
        elif numques=="3":
                print(f"""
                                Bye!! You go home with $ {prize1+ prize2}. Don't, uh, spend it all at once...""")
        elif numques=="4":
                print(f"""
                                Bye!! You go home with $ {prize1+ prize2+prize3}. Don't, uh, spend it all at once...""")
                exit(0)
        else:
                print("Invalid")
#-----------------------------------------------------------------------------------------------------
#Stage2

def game_start():    
    """In this function there is introduction to the game for the user and asks whether he wants to play"""
    
    print(
        """
                Well Well well congratulation on clearing the "Who is the fastest?" round.
                    
            As you know it is a gameshow to test how good is your general knowledge? This game 
            consists of 4 questions, you can easily  answer the questions. To help you out, there
            are "lifelines" that will let you either eliminate two of the four possible answers or 
            take a poll of the audience for what they think the right answer is. 

            Note: There is also a fifth option called  Walk away with money (If you guess incorrectly,
            you walk away with nothing) which will be available for some questions. But if you take a 
            lifeline for those questions you cannot use that (Walk away with money (If you guess 
            incorrectly, you walk away with nothing )) option

                                    ***Only one lifeline per question***.
                              Your host and friend for the game will be Computer
        
        """)
    time.sleep(2)
    global participant_name
    participant_name =None
    while participant_name == None or participant_name=="" or participant_name==" ":
        participant_name= input("\nPlease enter your name\n")      
    
    print("\nHello " + participant_name +", Welcome to the show: HOW GOOD IS YOUR GENERAL KNOWLEDGE!!!")
    nm_q= input("\nHow are you today?\n")
    if nm_q != "":
        print("\nVery well !! its time to make you little nervous \n")
    else:
        print("\nI don't think you want to answer this question. Lets proceed to actual brain testing questions\n")
    time.sleep(3)
    ready_to_play=input("Are you ready to play the game:\nPress 1 to play \nPress 2 if you dont want to play\n")
    
    #while loop will run till user inputs a valid input
    while (ready_to_play !="1" and ready_to_play !="2") and (ready_to_play.lower()!="one" and ready_to_play.lower()!="two") and (ready_to_play.lower()!="yes" and ready_to_play.lower()!="no"):
        ready_to_play=input("Please enter one of the following: Are you ready to play the game:\nType 1 to play \nType 2 if you dont want to play\n")
    
    if ready_to_play == "1" or ready_to_play.lower()=="one" or ready_to_play.lower()=="yes":
        print("\nOh yes!! I like that answer")
        print(
            """                          
                         Lets begin the show : HOW GOOD IS YOUR GENERAL KNOWLEDGE!!!
            
            """)
        time.sleep(3)
        question1()       
    else:
         print("""
         
                                             Bye!! It was nice hosting you.
             """)              


def question1():
    """Asking the user Question 1: There are only 4 options in this question"""
    print("Just a reminder : You have " + str(count_lifeline)+ " lifeline(s)\n") 
    print("*" * 60)
    print(f"Question1 for $ {prize1} \n What is three-fifths of 50?\n")
    time.sleep(2)
    choices=["1: 60", "2: 30", "3: 83.33", "4: 0.012" ]
    ans=choices[1]
    print(*choices, sep = "\n") 
    print("*" * 60)
    time.sleep(2)   
    set_var()
    final_answer=None
    final_answer=rep(answer,life_lines,not_blanks,choices,ans,"1")
    if final_answer=="2" or final_answer == "30" or final_answer.lower()=="thirty" or final_answer.lower()=="two":
        print('$ ' * 60, end='\n')
        print(f"""                   
                                        CONGRATULATIONS {participant_name}!!!
                                        YOU HAVE WON $ """+ str(prize1), end="\n")
        print("\n")
        print('$ ' * 60 + "\n\nDo you want continue playing?")
        continue_playing=input("TYPE 1 for Yes \n2 for No \n")
        quesnm="1"
        to_continue_play(quesnm,continue_playing)
    else:
        print("""
        
                                            You don't win anything
        """)
        fail()

#------------------------------------------------------------------------------------------------------------
#Stage3
#In stage 3, two questions are asked in both the questions fifth option is introduced
def question2():   
    """In this function fifth option is introduced"""
    
    print("Just a reminder : You have " + str(count_lifeline)+ " lifeline(s)\n") 
    print("*" * 60)
    print(f"Question2 for $ {prize2}\n During which year did World War I begin?\n")
    time.sleep(2)
    choices=["1: 1945", "2: 1920", "3: 1914", "4: 1898", "5: Walk away with $500 (If you guess incorrectly, you walk away with nothing)" ]
    ans=choices[2]
    print(*choices, sep = "\n") 
    print("*" * 60)
    time.sleep(2)   
    set_var()
    final_answer=None    
    final_answer=rep(answer,life_lines,not_blanks,choices,ans,"2")
    if final_answer=="3" or final_answer.lower()=="three" or final_answer=="1914":
        print('$ ' * 60, end='\n')
        print(f"""
                                            CONGRATULATIONS {participant_name}!!!
                                            YOU HAVE WON $ """+ str(prize1 + prize2), end="\n")
        print("\n")
        print('$ ' * 60 , end='\n')
        
        
        print(
            """             
                    Before we start playing again lets take a short break and when we come back 
                                there will be more questions and more knowledge
            
            """)
        for dot in range(3):
            print(
                """                                               .
                
                """)
            time.sleep(2)
        print(
            """
                    Wait... there is a question in break too for people who are sitting at home.
                    They can get a chance to skip the fastest round and enter the actual game....
                                        So the Question is.....
                                How many countries are in the world?
            
            """)
        print(
            """                             
                        Please send your answers to us.. Type SMS (your answer) to 242424
            
            """)
        input("<Press Enter to Resume from the break>")
        print(
              """                     
                                    Welcome back from the break!!!!
              
              """)
        profession=input(f"\nWhat is your profession {participant_name}?")
        if profession=="":
            print("Ok! I understand you don't want to answer\n")
        else:
            print("Oh nice.. Lets not waste any more time\n")
        
        continue_playing=input("\nDo you want to continue playing?\nTYPE 1 for Yes \n2 for No \n")
        quesnm="2"
        to_continue_play(quesnm,continue_playing)
        
    elif final_answer=="5" or final_answer.lower()=="five":
        print(f"""
        
                                                Bye!! You go home with $ {prize1} 
            """)
    else:
        print(f"""
        
                                            You had ${prize1} but went home with nothing
                                
             """)
        fail()

def question3(): 
    """In this function fifth option is introduced"""
    print("Just a reminder : You have " + str(count_lifeline)+ " lifeline(s)\n") 
    time.sleep(2)
    print("*" * 40)
    print(f"Question3 for $ {prize3}\nIn medieval times, which of these professionals was responsible for entertaining the royal courts?\n")
    time.sleep(2)
    choices=["1: Barber", "2: Knight", "3: Blacksmith", "4: Jester" ,"5: Walk away with $1500 (If you guess incorrectly, you walk away with nothing)" ]
    ans=choices[3]
    print(*choices, sep = "\n") 
    print("*" * 40)
    time.sleep(2)   
    set_var()
    final_answer=None    
    final_answer=rep(answer,life_lines,not_blanks,choices,ans,"3")
    if final_answer=="4" or final_answer.lower()=="four" or final_answer.lower()=="jester":
        print(f"""          
                    
                    
                       {participant_name}......{participant_name} You could have gone home with $1500..... 
                                                      .
                                                      .
                                                      .
                                                
                                            but no need to worry!!
              
              """)
        
        print("\n")
        print('$ ' * 60, end='\n')
        print(f"""            
                                        YOU HAVE WON $ """+ str(prize1 + prize2 + prize3), end="\n")
        print("\n")        
        print('$ ' * 60)
        from_participant=input("\nIt feels good right?")
        print("\nThis is fun..This is fun game for you!!")
        input("\nSo tell me what will you do with the money?")
        print("\nAll right buddy you are doing great.. Lets continue\n\nDo you want continue playing?")
        continue_playing=input("TYPE 1 for Yes \n2 for No \n")
        quesnm="3"
        to_continue_play(quesnm,continue_playing)
    elif final_answer=="5" or final_answer.lower()=="five":
        print(f"""
        
                    Bye!! You go home with $ {prize1+ prize2}.Don't, uh, spend it all at once...
        
             """)
    else:
        print(f"""
        
                    You could have won $ {prize1 + prize2+ prize3} but now you go home with nothing
        
               """)
        fail()

#---------------------------------------------------------------------------------------------------------------
#Stage4

def questionLast(): 
    """In the last question there is no fifth option and no lifeline for the user"""
    
    print("-" * 100)
    print(
            """
                    In this last question you will not get any lifeline. Moreover, you will
                           also not get the fifth option of taking the money home.
                    
                         So if you answer incorrectly you will lose all your money
                    
            """) 
    print("Do you want to continue playing?")
    continue_playing=input("TYPE 1 for Yes \n2 for No \n")
    quesnm="4"
    to_continue_play(quesnm,continue_playing)
    print("-" * 100)
    print("\n")
    print("*" * 60)
    print(f"Last Question for $ {prize4} \nWhich insect shorted out an early supercomputer and inspired the term 'computer bug'?\n")
    choices=["1: Moth", "2: Roach", "3: Fly", "4: Beetle"]
    print(*choices, sep = "\n") 
    print("*" * 60) 
    final_answer=None
    final_answer=input("Please give an answer")
    final_answer= check_answers_inlist(quesnm,final_answer)
    if final_answer=="1" or final_answer.lower()=="one" or final_answer.lower()=="moth":
        print("\n")
        print('$ ' * 60, end='\n')
        congrats()
        print(
                """ 
                       That was impressive.. That was really impressive.. You win this show..
                
                """)
        print(f"""   
                                            YOU TAKE HOME $ """+ str(prize1 + prize2 + prize3+ prize4), end="\n")
        print("\n")
        print('$ ' * 60, end='\n')
        print(f"""              
                                  Wait Wait Wait its not over {participant_name}... 
                                  There is one surprise waiting for you.
               
               """)
        time.sleep(2)
        favour_round()
    else:
        print(
                """
                                        You lost it! You don't take anything home
                
                """)
        fail()              

#----------------------------------------------------------------------------------------------------------------
#Stage5

def favour_round():
    """In this function the winner has to choose a color and randomly a name will get associated with 
    that color who will come as the next contestant for actual game skipping the 'who is the fastest' round    
    """
    
    print(
            """                   
                        Here is the time for you to give back something to the game
                                    No No we are not asking for money..
                  You get a chance to pick a contestant who will skip the round 'who is the fastest?' 
                             and proceed to the actual game for the next week..

                          This is exclusive for someone who wins the show
            
            """)

    print(
            """           
                            So are you ready to return back the favour?
            
            """)
    input("""       <Press Enter to continue>""")
    
    for fullstop in range(4):
        print(f"""
                                                    .""",end="\n")
        time.sleep(1)
    print(
            """                    
                
                By the way the answer for question which was asked in the break is : 195 countries..
                      So here is the list of contestants who won the commercial break round:
                                             Ann George
                                               Deb B
                                               Yi Tan
                                               Tintin
                                               Dexter
          
            """)
    li_commercial_colors=['Blue','Pink','Yellow','Red','Orange']
    li_commercial_participants=['Ann George','Deb B','Yi Tan','Tintin','Dexter']
    input("<Please Enter to continue>")
    print("\nPlease enter a hat color. The hat color you choose can change destiny of that person")
    print("\nYou get only one chance and if you choose a hat color which is incorrect then those contestants will lose the opportunity to come to the show")
    print('* '* 40 )
    print(*li_commercial_colors, sep = "\n")
    print('* '* 40 )
    choice=input("\nEnter a color\n")
    
    
    name=''
    random.shuffle(li_commercial_participants)
    
    #for loop over two list so that if the color matches with the user input participants name will also get printed
    for color, name_partcipant in zip(li_commercial_colors,li_commercial_participants):
        if color.lower() ==choice.lower(): 
            #print(color+ " - " +name_partcipant)
            name=name_partcipant
            break  
    print("\nLets see who is the winner\n")
    #for loop to print loading
    for x in range (0,5):  
        dot = "Loading" + "." * x
        print (dot, end="\r")
        time.sleep(1)
    time.sleep(2)
    if name=="":
        print("""
                                    No one gets the opportunity to come. Better luck next time.
        
        """)
    else:
        
        print(f"""

                        Congratulations {name}.. You will be the next person coming to this show.

              """)  
        print("""

                                    Till then good bye and learn more

              """)
#----------------------------------------------------------------------------------------------------------------    
#Stage 1    
def filtering_contestant():
    """In this function if the user inputs all the answers below 20 seconds and if the 
    answers are right then user advances to next level(stage) 
    """
    
    
    print(
        """ 
                    Hello participants!!!! Welcome to the show: HOW GOOD IS YOUR GENERAL KNOWLEDGE?
                    
                  You are one of the 10 participants to reach this level in the gameshow. To proceed to the 
                 advance levels, participants have to go through the fastest round. In this round the person
                 who answers fastest will get a chance to play the advance levels where the actual game begins.
        
        """)
    filter_play=input("Are you ready to play 'Who is the fastest' round?\nPress 1 to play \nPress 2 if you dont want to play\n")
     #while loop will run till user inputs a valid input
    while (filter_play !="1" and filter_play !="2") and (filter_play.lower()!="one" and filter_play.lower()!="two") and (filter_play.lower()!="yes" and filter_play.lower()!="no"):
        filter_play=input("Please enter one of the following: Are you ready to play who is the fastest round?:\nType 1 to play \nType 2 if you dont want to play\n")
    if filter_play == "1" or filter_play.lower()=="one" or filter_play.lower()=="yes":
        print("\nLets begin the round : Who is the fastest?\n")
        time.sleep(3)
        print(f"Arrange the planets in the order in which they appear\n")
        time.sleep(2)
        filterquest=["1: Mars", "2: Mercury", "3: Earth", "4: Venus" ]
        print(*filterquest, sep = "\n") 
        startTime = time.time()
        f1=input("\nFirst planet close to the sun \n")
        s1=input("Second planet close to the sun \n")
        t1=input("Third planet close to the sun \n")
        f4=input("Fourth planet close to the sun \n")
        endTime = time.time()
        total_time=round(endTime - startTime) 
        if (f1.lower()=="mercury" or f1=="2" or f1.lower()=="two") and (s1.lower()=="venus" or s1=="4" or s1.lower()=="four") and (t1.lower()=="earth" or t1=="3" or t1.lower()=="three") and (f4.lower()=="mars" or f4=="1" or f4.lower()=="one"):
                if total_time < 20:  
                    congrats()
                    print(f"You took {total_time} seconds least time among all participants!!!! You get a chance to proceed to the advance level.")
                    print('- '*40)
                    input("<Press Enter to continue>")
                    game_start()
                else:
                    print(f"""
                            Sorry you cannot proceed to the other round(you took more time than others).
                            
                            """)
        else:
            print("""
                            Sorry you cannot proceed to the other round(your answer is wrong).
                 """)
    else:
         print(
               """
                                           Bye!! It was nice hosting you
                
               """)  
    
    
filtering_contestant()
```
